```python
t = int(input())
for _ in range(t):
    n = int(input())
    num = [num for num in input().split()]
    num_dct = {num[i]: 0 for i in range(0, len(num))}
    
    print(num)
    
    flag = 0
    
    for i in num:
        num_dct[i] += 1
    
    print(num_dct)
    
    for key, value in num_dct.items():
        if value % 2 == 1:
            flag = 1
            print(key)
            break
    
            
    if flag == 0:
        print(0)
    
```
